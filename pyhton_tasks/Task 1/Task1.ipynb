```python
import pandas as pd
import numpy as np

energy = pd.read_excel ('data/Energy Indicators.xls',skiprows=16,skipfooter=38,usecols="C:F",na_values="...")
energy.drop([0],axis=0,inplace=True)

energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy[['Energy Supply', 'Energy Supply per Capita', '% Renewable']] = energy[['Energy Supply', 'Energy Supply per Capita', '% Renewable']].apply(pd.to_numeric)
```

```python
energy.head(10)
```

```python
energy.dtypes
```

```python
#convert energy supply column
energy['Energy Supply']*=1000000
energy.head(10)
```

```python
#remove numbers in the end of string
energy['Country'] = energy['Country'].str.replace("[0-9]+$", "")
#remove parentheses
energy['Country'] = energy['Country'].str.replace("\(.*\)", "")

energy['Country'].replace({"Republic of Korea": "South Korea","United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"},inplace=True)
```

```python
energy['Country'].head(5)
```

```python
boolean = energy.Country == "United Kingdom"

boll = pd.Series(boolean)
energy[boll]
```

```python
GDP = pd.read_csv('data/world_bank.csv',skiprows = 4)
GDP = GDP.rename(columns={'Country Name': 'Country'})
```

```python
GDP.head(5)
```

```python
GDP['Country'].replace({"Korea, Rep.": "South Korea", 
"Iran, Islamic Rep.": "Iran","Hong Kong SAR, China": "Hong Kong"},inplace=True)
GDP = GDP.dropna(axis=1, how='all')
```

### Step 5

```python
Top15ScimEn = ScimEn.sort_values(by="Rank", ascending=True).head(15)

GDPandScimEn = pd.merge(Top15ScimEn,GDP[['Country','2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']],
on='Country')
GDPandScimEn.head(15)
# GDPandScimEn.shape
```

```python
GDPandScimEn['avgGDP'] = GDPandScimEn[['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']].mean(axis = 1)
# GDPandScimEn['avgGDP']
GDPandScimEn[['Country','avgGDP']].sort_values(by = 'avgGDP', ascending=False).head(15)
```

## step 6

```python
TOP15energyANDScimEn = pd.merge(Top15ScimEn,energy,on='Country')
TOP15energyANDScimEn.set_index('Country', inplace = True)
# TOP15energyANDScimEn.dtypes
```

```python
mean_energy_per_captia = TOP15energyANDScimEn['Energy Supply per Capita'].mean()
print(mean_energy_per_captia)
```

## step 7

```python
country_with_max_Renewable = TOP15energyANDScimEn['% Renewable'].idxmax()
maximum_Renewable = (country_with_max_Renewable,TOP15energyANDScimEn['% Renewable'].max())
print(maximum_Renewable)
```

## step 8

```python
TOP15energyANDScimEn['ratio'] = TOP15energyANDScimEn['Self-citations'] / TOP15energyANDScimEn['Citations']

country_with_max_ratio = TOP15energyANDScimEn['ratio'].idxmax()
maximum_ratio = (country_with_max_ratio,TOP15energyANDScimEn['ratio'].max())
print(maximum_ratio)
```

## step 9

```python
median = TOP15energyANDScimEn['% Renewable'].median()

TOP15energyANDScimEn['HighRenew'] = TOP15energyANDScimEn['% Renewable'].apply(lambda x : 1 if x > median else 0 )
# TOP15energyANDScimEn.head(15)

#HighRenew series
TOP15energyANDScimEn['HighRenew']
```

## step 10

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```

```python
TOP15energyANDScimEn = TOP15energyANDScimEn.reset_index()
TOP15energyANDScimEn['Continent'] = TOP15energyANDScimEn['Country'].map(ContinentDict)
TOP15energyANDScimEn['estimated_population'] = TOP15energyANDScimEn['Energy Supply'] / TOP15energyANDScimEn['Energy Supply per Capita']
TOP15energyANDScimEn.head()
```

```python
aggregateDF = TOP15energyANDScimEn.loc [:,['Continent','estimated_population']]

aggregateDF = aggregateDF.groupby('Continent').agg([np.size,np.sum,np.mean,np.std])
aggregateDF.head()
```
